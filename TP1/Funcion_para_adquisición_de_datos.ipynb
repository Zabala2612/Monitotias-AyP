## <font color='46B8A9'> **Función cut()**

Esta función se utiliza para segmentar y agrupar datos continuos en categorías o "bins", lo que resulta muy útil cuando se quieren analizar datos distribuidos en rangos. Esto permite dividir un conjunto de datos numéricos en intervalos que facilitan la interpretación y análisis de patrones.

### <font color='8EC044'> **Ejemplo**

Supongamos que tenemos un DataFrame con las edades de varias personas y queremos agruparlas en diferentes rangos de edad, como "Joven", "Adulto" y "Mayor".

```python
import pandas as pd

# Lista de edades
data = {'nombre': ['Felipe', 'Mateo', 'Luis', 'Santiago', 'Pedro'],
        'edad': [23, 42, 68, 34, 15]}

df = pd.DataFrame(data)

print("DataFrame original:\n")
df
```

```python
# Usamos pd.cut para agrupar las edades en categorías
bins = [0, 18, 35, 65, 100]  # Definimos los intervalos de edad
labels = ['Menor', 'Joven', 'Adulto', 'Mayor']  # Etiquetas para cada intervalo

df['grupo_edad'] = pd.cut(df['edad'], bins=bins, labels=labels)
"""
La función cut() toma los valores de la columna 'edad' y los agrupa en los
intervalos definidos por bins. Luego, asigna la etiqueta correspondiente
de acuerdo con el rango en el que cae cada valor.
"""

print("""El resultado es una nueva columna llamada 'grupo_edad' en el DataFrame,
que contiene las etiquetas correspondientes a cada rango de edad:\n""")
df
```
